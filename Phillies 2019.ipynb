# Phillies 2019

```python
url4 = 'https://www.baseball-reference.com/teams/tgl.cgi?team=PHI&t=b&year=2019'
page4 = requests.get(url4)
soup4 = BeautifulSoup(page4.text, 'html.parser')

def get_stat2(stat):
    statistic = soup4.find_all('td', attrs = {'data-stat': stat})
    Stat = []
    for x in statistic:
        Stat.append(str(x))
    final_stat = []
    for x in range(len(Stat)):
        bg = re.compile(r'>(.?\d+)</td>')
        lemmon =bg.findall(Stat[x])
        final_stat.append(lemmon)
    final_stat = list(chain.from_iterable(final_stat))
    return final_stat


PHI_2019_game_num = get_stat2('team_game_num')
PHI_2019_PA = get_stat2('PA')
PHI_2019_AB = get_stat2('AB')
PHI_2019_R = get_stat2('R')
PHI_2019_H = get_stat2('H')
PHI_2019_2B = get_stat2('2B')
PHI_2019_3B = get_stat2('3B')
PHI_2019_HR = get_stat2('HR')
PHI_2019_RBI = get_stat2('RBI')
PHI_2019_BB = get_stat2('BB')
PHI_2019_SO = get_stat2('SO')
PHI_2019_SB = get_stat2('SB')
PHI_2019_CS = get_stat2('CS')
PHI_2019_BA = get_stat2('batting_avg')
PHI_2019_OBP = get_stat2('onbase_perc')
PHI_2019_SLG = get_stat2('slugging_perc')
PHI_2019_OPS = get_stat2('onbase_plus_slugging')
PHI_2019_LOB = get_stat2('LOB')
phillies_2019_offense = pd.DataFrame({
                                  'PA': PHI_2019_PA,
                                  'AB': PHI_2019_AB,
                                  'R':  PHI_2019_R,
                                  'H':  PHI_2019_H,
                                  '2B': PHI_2019_2B,
                                  '3B': PHI_2019_3B, 
                                  'HR': PHI_2019_HR, 
                                  'RBI':PHI_2019_RBI,
                                  'BB': PHI_2019_BB,
                                  'SO': PHI_2019_SO,
                                  'SB': PHI_2019_SB,
                                  'CS': PHI_2019_CS,
                                  'LOB':PHI_2019_LOB})

url5 = 'https://www.baseball-reference.com/teams/tgl.cgi?team=PHI&t=p&year=2019'
page5 = requests.get(url5)
soup5 = BeautifulSoup(page5.text, 'html.parser')

def get_stat3(stat):
    statistic = soup5.find_all('td', attrs = {'data-stat': stat})
    Stat = []
    for x in statistic:
        Stat.append(str(x))
    final_stat = []
    for x in range(len(Stat)):
        bg = re.compile(r'>(.+)</td>')
        lemmon =bg.findall(Stat[x])
        final_stat.append(lemmon)
    final_stat = list(chain.from_iterable(final_stat))
    return final_stat

PHI_2019_P_game_num = get_stat3('team_game_num')
PHI_2019_P_IP = get_stat3('IP')
PHI_2019_P_H = get_stat3('H')
PHI_2019_P_R = get_stat3('R')
PHI_2019_P_ER = get_stat3('ER')
PHI_2019_P_UER = get_stat3('UER')
PHI_2019_P_BB = get_stat3('BB')
PHI_2019_P_SO = get_stat3('SO')
PHI_2019_P_HR = get_stat3('HR')
PHI_2019_P_HBP = get_stat3('HBP')
PHI_2019_P_ERA = get_stat3('earned_run_avg')

phillies_2019_pitching = pd.DataFrame({'Game':  PHI_2019_P_game_num, 
                                         'IP':  PHI_2019_P_IP, 
                                         'H':   PHI_2019_P_H,
                                         'R':   PHI_2019_P_R,
                                         'ER':  PHI_2019_P_ER,
                                         'UER': PHI_2019_P_UER,
                                         'BB':  PHI_2019_P_BB,
                                         'SO':  PHI_2019_P_SO,
                                         'HR':  PHI_2019_P_HR,
                                         'HBP': PHI_2019_P_HBP})

# phillies_offense = phillies_offense.drop(columns = ['Game'])

phillies_2019_pitching = phillies_2019_pitching.rename(columns = {'H': 'Hits', 'R':'Runs', 'BB':'Walks', 'SO':'Strikeouts', 'HR': 'HR_Allowed'})

frames = [phillies_2019_pitching, phillies_2019_offense]

phillies_2019 = pd.concat(frames, axis=1)

phillies_2019['Win'] =''

columns_int = ['Hits', 'Runs', 'ER', 'UER', 'Walks','Strikeouts', 'HR_Allowed', 'HBP', 'PA', 'AB', 'R', 'H', '2B','3B', 'HR', 'RBI', 'BB', 'SO', 'SB', 'CS', 'LOB']

columns_float = ['IP']

for x in columns_int:
    for y in range(len(phillies_2019[x])):
        phillies_2019.loc[y,x] = int(phillies_2019.loc[y, x])

for x in columns_float:
    for y in range(len(phillies_2019[x])):
        phillies_2019.loc[y,x] = float(phillies_2019.loc[y, x])  

for x in range(len(phillies_2019['Game'])):
    if phillies_2019.loc[x, 'Runs'] > phillies_2019.loc[x, 'R']:
        phillies_2019.loc[x, 'Win'] = 0
    else:
        phillies_2019.loc[x, 'Win'] = 1
```
